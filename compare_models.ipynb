```python
# Model Comparison - ViT, ResNet, DenseNet, EfficientNet

import os
import torch
import torch.nn as nn
import numpy as np
from torchvision import models, transforms
from torch.utils.data import Dataset, DataLoader, Subset
from PIL import Image, ImageFile
import timm
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, roc_curve, f1_score, accuracy_score
from sklearn.model_selection import train_test_split
from collections import OrderedDict
import random

ImageFile.LOAD_TRUNCATED_IMAGES = True
Image.MAX_IMAGE_PIXELS = None

# Dataset definition
class HistoDataset(Dataset):
    def __init__(self, root_dir, transform=None, max_samples=10000):
        self.classes = ['normal', 'tumor']
        self.samples = []
        self.transform = transform

        for cls in self.classes:
            cls_dir = os.path.join(root_dir, cls)
            all_files = sorted(os.listdir(cls_dir))
            for f in all_files:
                path = os.path.join(cls_dir, f)
                if not (f.endswith('.png') and os.path.getsize(path) > 1024):
                    continue
                try:
                    # Try opening and converting image once
                    with Image.open(path) as img:
                        img.convert('RGB')
                    self.samples.append((path, self.classes.index(cls)))
                except Exception as e:
                    print(f"Skipping corrupted image: {path}\nError: {e}")

        random.shuffle(self.samples)
        self.samples = self.samples[:max_samples]

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, idx):
        path, label = self.samples[idx]
        try:
            img = Image.open(path).convert('RGB').resize((224, 224))
            if self.transform:
                img = self.transform(img)
            return img, label
        except Exception as e:
            print(f"Error reading {path}: {e}")
            return torch.zeros(3, 224, 224), -1

# Utility: load model weights safely
def safe_load_weights(model, checkpoint_path, device):
    checkpoint = torch.load(checkpoint_path, map_location=device)
    state_dict = checkpoint.get('model', checkpoint)
    new_state_dict = OrderedDict((k.replace("model.", ""), v) for k, v in state_dict.items())
    model.load_state_dict({k: v for k, v in new_state_dict.items() if k in model.state_dict()}, strict=False)
    return model

# Evaluation loop
def evaluate(model, dataloader, device):
    model.eval()
    all_preds, all_targets, all_probs = [], [], []
    with torch.no_grad():
        for x, y in dataloader:
            x, y = x.to(device), y.to(device)
            out = model(x)
            probs = torch.softmax(out, dim=1)[:, 1]  # Probability of class 1 (tumor)
            preds = torch.argmax(out, dim=1)
            all_preds.extend(preds.cpu().numpy())
            all_targets.extend(y.cpu().numpy())
            all_probs.extend(probs.cpu().numpy())
    return np.array(all_preds), np.array(all_targets), np.array(all_probs)

# Visualization helpers
def plot_conf_matrix(y_true, y_pred, title):
    cm = confusion_matrix(y_true, y_pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=["Normal", "Tumor"], yticklabels=["Normal", "Tumor"])
    plt.title(title)
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.show()

def plot_roc_curves(roc_data):
    plt.figure(figsize=(8, 6))
    for model_name, (fpr, tpr, auc) in roc_data.items():
        plt.plot(fpr, tpr, label=f'{model_name} (AUROC = {auc:.2f})')
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title("ROC Curves")
    plt.legend()
    plt.grid(True)
    plt.show()

# Main comparison script
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
DATA_DIR = "processed_slides"
BATCH_SIZE = 32
val_transform = transforms.Compose([transforms.ToTensor()])

# Prepare val dataloader
dataset = HistoDataset(DATA_DIR, transform=val_transform)
labels = [s[1] for s in dataset.samples]
_, val_idx = train_test_split(range(len(dataset)), test_size=0.2, stratify=labels)
val_loader = DataLoader(Subset(dataset, val_idx), batch_size=BATCH_SIZE)

# Define models and checkpoints
model_configs = {
    "ViT": ("out_model/uterine_cancer_vit.pth", lambda: timm.create_model('vit_base_patch16_224', pretrained=False, num_classes=2)),
    "ResNet34": ("out_model/uterine_cancer_resnet34.pth", lambda: models.resnet34(pretrained=False, num_classes=2)),
    "EfficientNet": ("out_model/uterine_cancer_efficientnet.pth", lambda: timm.create_model('efficientnet_b0', pretrained=False, num_classes=2)),
    "DenseNet121": ("out_model/uterine_cancer_densenet.pth", lambda: models.densenet121(pretrained=False, num_classes=2)),
}

# Store evaluation results
metrics_table = []
roc_data = {}

for model_name, (ckpt, model_fn) in model_configs.items():
    model = model_fn().to(DEVICE)
    safe_load_weights(model, ckpt, DEVICE)
    preds, targets, probs = evaluate(model, val_loader, DEVICE)

    acc = accuracy_score(targets, preds)
    f1 = f1_score(targets, preds)
    auc = roc_auc_score(targets, probs)
    fpr, tpr, _ = roc_curve(targets, probs)

    metrics_table.append((model_name, acc, f1, auc))
    roc_data[model_name] = (fpr, tpr, auc)

    print(f"\n[{model_name} Classification Report]")
    print(classification_report(targets, preds, target_names=["Normal", "Tumor"]))
    plot_conf_matrix(targets, preds, f"{model_name} Confusion Matrix")

# Summary Table
import pandas as pd
df_metrics = pd.DataFrame(metrics_table, columns=["Model", "Accuracy", "F1 Score", "AUROC"])
display(df_metrics)

# Plot ROC curves
plot_roc_curves(roc_data)
```

```python
from sklearn.metrics import precision_recall_curve, average_precision_score

def plot_precision_recall_curves(pr_data):
    plt.figure(figsize=(8, 6))
    for model_name, (precision, recall, ap) in pr_data.items():
        plt.plot(recall, precision, label=f'{model_name} (AP = {ap:.2f})')
    plt.xlabel("Recall")
    plt.ylabel("Precision")
    plt.title("Precision-Recall Curves")
    plt.legend()
    plt.grid(True)
    plt.show()

pr_data = {}
for model_name, (ckpt, model_fn) in model_configs.items():
    model = model_fn().to(DEVICE)
    safe_load_weights(model, ckpt, DEVICE)
    preds, targets, probs = evaluate(model, val_loader, DEVICE)

    precision, recall, _ = precision_recall_curve(targets, probs)
    ap = average_precision_score(targets, probs)
    pr_data[model_name] = (precision, recall, ap)

plot_precision_recall_curves(pr_data)
```

```python
def calculate_specificity(y_true, y_pred):
    cm = confusion_matrix(y_true, y_pred)
    tn = cm[0][0]
    fp = cm[0][1]
    return tn / (tn + fp)

for model_name, (ckpt, model_fn) in model_configs.items():
    model = model_fn().to(DEVICE)
    safe_load_weights(model, ckpt, DEVICE)
    preds, targets, probs = evaluate(model, val_loader, DEVICE)

    specificity = calculate_specificity(targets, preds)
    print(f"{model_name} Specificity: {specificity:.2f}")
```

```python
from sklearn.metrics import log_loss

# Function to calculate log loss for each model
def calculate_log_loss(y_true, y_probs):
    return log_loss(y_true, y_probs)

log_loss_data = {}

for model_name, (ckpt, model_fn) in model_configs.items():
    model = model_fn().to(DEVICE)
    safe_load_weights(model, ckpt, DEVICE)
    preds, targets, probs = evaluate(model, val_loader, DEVICE)

    # Calculate log loss
    loss = calculate_log_loss(targets, probs)
    log_loss_data[model_name] = loss

# Display log loss for each model
for model_name, loss in log_loss_data.items():
    print(f"{model_name} Log Loss: {loss:.4f}")
```
